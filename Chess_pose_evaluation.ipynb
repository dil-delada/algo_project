```python
def evaluate(board):
    white_score = 0
    black_score = 0
    for i in range(8):
        for j in range(8):
            if board[i][j] == 11:
                if j <= 5:
                    black_score += 10
                if j == 1:
                    black_score += 30
                black_score += 100
            if board[i][j] == 12:
                if i < 1 or i > 6 or j < 1 or j > 6:
                    black_score -= 30
                black_score += 300
            if board[i][j] == 14:
                black_score += 350
            if board[i][j] == 15:
                black_score += 500
            if board[i][j] == 18:
                black_score += 900
                
            if board[i][j] == 1:
                if j >= 4:
                    white_score += 10
                if j == 7:
                    white_score += 30
                white_score += 100
            if board[i][j] == 2:
                if i < 1 or i > 6 or j < 1 or j > 6:
                    white_score -= 30
                white_score += 300
            if board[i][j] == 4:
                white_score += 350
            if board[i][j] == 5:
                white_score += 500
            if board[i][j] == 8:
                white_score += 900
    return (white_score - black_score) / white_score
```

```python
def all_possible_moves(board, Player):
    moves = []
    diap = np.arange(8)
    players = [10 * Player + 1, 10 * Player + 2, 10 * Player + 4, 10 * Player + 5, 10 * Player + 8, 10 * Player + 9]
    for i in range(8):
        for j in range(8):
            if board[i][j] == 10 * Player + 1:
                if j + 1 - 2 * Player >= 0 and j + 1 - 2 * Player <= 7 and board[i][j + 1 - 2 * Player] == 0:
                    moves.append(chr(ord('a') + i) + str(j + 2 - 2 * Player))
                    if j + 2 - 2 * Player >= 0 and j + 2 - 2 * Player <= 7 and board[i][j + 2 - 2 * Player] == 0:
                        if j == 1 + 6 * Player:
                            moves.append(chr(ord('a') + i) + str(j + 3 - 2 * Player))
            if board[i][j] == 10 * Player + 2:
                moves = Knight_moves(board, moves, i, j, diap, players)
            if board[i][j] == 10 * Player + 4:
                moves = Bishop_moves(board, moves, players, i, j)
            if board[i][j] == 10 * Player + 5:
                k = 1
                while i + k <= 7:
                    if board[i + k][j] not in players:
                        if board[i + k][j] == 0:
                            moves.append("R" + chr(ord('a') + i + k) + str(j + 1))
                        else:
                            moves.append("Rx" + chr(ord('a') + i + k) + str(j + 1))
                    else:
                        break
                    k += 1
                k = 1
                while i - k >= 0:
                    if board[i - k][j] not in players:
                        if board[i - k][j] == 0:
                            moves.append("R" + chr(ord('a') + i - k) + str(j + 1))
                        else:
                            moves.append("Rx" + chr(ord('a') + i - k) + str(j + 1))
                    else:
                        break
                    k += 1
                k = 1
                while j - k >= 0:
                    if board[i][j - k] not in players:
                        if board[i][j - k] == 0:
                            moves.append("R" + chr(ord('a') + i) + str(j - k + 1))
                        else:
                            moves.append("Rx" + chr(ord('a') + i) + str(j - k + 1))
                    else:
                        break
                    k += 1
                k = 1
                while j + k <= 7:
                    if board[i][j + k] not in players:
                        if board[i][j + k] == 0:
                            moves.append("R" + chr(ord('a') + i) + str(j + k + 1))
                        else:
                            moves.append("Rx" + chr(ord('a') + i) + str(j + k + 1))
                    else:
                        break
                    k += 1
            if board[i][j] == 10 * Player + 8:
                k = 1
            if board[i][j] == 10 * Player + 9:
                if i + 1 <= 7:
                    if board[i + 1][j] not in players:
                        if board[i + 1][j] == 0:
                            moves.append("K" + chr(ord('a') + i + 1) + str(j + 1))
                        else:
                            moves.append("Kx" + chr(ord('a') + i + 1) + str(j + 1))
                    if j + 1 <= 7:
                        if board[i + 1][j + 1] not in players:
                            if board[i + 1][j + 1] == 0:
                                moves.append("K" + chr(ord('a') + i + 1) + str(j + 2))
                            else:
                                moves.append("Kx" + chr(ord('a') + i + 1) + str(j + 2))
                    if j - 1 >= 0:
                        if board[i + 1][j - 1] not in players:
                            if board[i + 1][j - 1] == 0:
                                moves.append("K" + chr(ord('a') + i + 1) + str(j))
                            else:
                                moves.append("Kx" + chr(ord('a') + i + 1) + str(j))
                                
                if i - 1 >= 0:
                    if board[i + 1][j] not in players:
                        if board[i + 1][j] == 0:
                            moves.append("K" + chr(ord('a') + i + 1) + str(j + 1))
                        else:
                            moves.append("Kx" + chr(ord('a') + i + 1) + str(j + 1))
                    if j + 1 <= 7:
                        if board[i + 1][j + 1] not in players:
                            if board[i + 1][j + 1] == 0:
                                moves.append("K" + chr(ord('a') + i + 1) + str(j + 2))
                            else:
                                moves.append("Kx" + chr(ord('a') + i + 1) + str(j + 2))
                    if j - 1 >= 0:
                        if board[i + 1][j - 1] not in players:
                            if board[i + 1][j - 1] == 0:
                                moves.append("K" + chr(ord('a') + i + 1) + str(j))
                            else:
                                moves.append("Kx" + chr(ord('a') + i + 1) + str(j))
    return moves
```
